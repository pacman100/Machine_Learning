```python
import numpy as np
#import sklearn as sk
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import os
import tensorflow as tf
import csv
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import confusion_matrix
```

```python
print(tf.__version__)
#print(sk.__version__)
print(np.__version__)
```

```python
x_train = []
y_train = []
x_valid = []
y_valid = []
x_test = []
y_test = []

with open('./group8/class1_train.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_train.append([float(a), float(b)])
        y_train.append(1)

with open('./group8/class2_train.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_train.append([float(a), float(b)])
        y_train.append(2)
        


with open('./group8/class1_val.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_valid.append([float(a), float(b)])
        y_valid.append(1)

with open('./group8/class2_val.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_valid.append([float(a), float(b)])
        y_valid.append(2)
    
        
        
with open('./group8/class1_test.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_test.append([float(a), float(b)])
        y_test.append(1)

with open('./group8/class2_test.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_test.append([float(a), float(b)])
        y_test.append(2)

        

x_train = np.array(x_train)
y_train = np.array(y_train)
x_valid = np.array(x_valid)
y_valid = np.array(y_valid)
x_test = np.array(x_test)
y_test = np.array(y_test)

plt.plot(x_train[:,0], x_train[:,1], 'ro')
plt.xlabel('X1 Value')
plt.ylabel('X2 Value')
plt.title('Linearly Non-Separable Training Dataset (Group-8)')
plt.show()
```

```python
clf = SVC()
clf.fit(x_train, y_train)
```

```python
y_pred_train = clf.predict(x_train)
acc = clf.score(x_train, y_train)
print("The accuracy of the Non-Linear SVM model on the training data:", acc*100,"%")

plot_decision_regions(x_train, y_train, clf=clf, legend=3)
plt.show()

#confusion Matrix
confusion_matrix(y_train, y_pred_train,labels=[1, 2])
```

```python
y_pred_valid = clf.predict(x_valid)
acc = clf.score(x_valid, y_valid)
print("The accuracy of the Non-Linear SVM model on the validation data:", acc*100,"%")

plot_decision_regions(x_valid, y_valid, clf=clf, legend=3)
plt.show()

#confusion Matrix
confusion_matrix(y_valid, y_pred_valid,labels=[1, 2])
```

```python
y_pred_test = clf.predict(x_test)
acc = clf.score(x_test, y_test)
print("The accuracy of the Non-Linear SVM model on the Test data:", acc*100,"%")

plot_decision_regions(x_test, y_test, clf=clf, legend=3)
plt.show()

#confusion Matrix
confusion_matrix(y_test, y_pred_test,labels=[1, 2])
```
