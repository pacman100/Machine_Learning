```python
import numpy as np
#import sklearn as sk
from sklearn import linear_model
import matplotlib.pyplot as plt
import os
import tensorflow as tf
import csv
```

```python
print(tf.__version__)
#print(sk.__version__)
print(np.__version__)
```

```python
x_train = []
y_train = []
x_valid = []
y_valid = []
x_test = []
y_test = []

with open('group8_train.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_train.append(float(a))
        y_train.append(float(b))

with open('group8_val.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_valid.append(float(a))
        y_valid.append(float(b))
        
with open('group8_test.txt', 'r') as file:
    for row in file:
        a,b = row.split()
        x_test.append(float(a))
        y_test.append(float(b))
        
x_train = np.array(x_train)
y_train = np.array(y_train)
x_valid = np.array(x_valid)
y_valid = np.array(y_valid)
x_test = np.array(x_test)
y_test = np.array(y_test)

train_feature_1 = (x_train - np.mean(x_train))/np.std(x_train)
x_train2 = np.power(x_train, 2)
train_feature_2 = (x_train2 - np.mean(x_train2))/np.std(x_train2)
val_feature_1 = (x_valid - np.mean(x_valid))/np.std(x_valid)
x_valid2 = np.power(x_valid, 2)
val_feature_2 = (x_valid2 - np.mean(x_valid2))/np.std(x_valid2)
test_feature_1 = (x_test - np.mean(x_test))/np.std(x_test)
x_test2 = np.power(x_test, 2)
test_feature_2 = (x_test2 - np.mean(x_test2))/np.std(x_test2)


x_train_features = np.hstack((train_feature_1.reshape(-1,1), train_feature_2.reshape(-1,1)))
#print(x_train_features[0:10, :])
y_train_features = y_train
x_valid_features = np.hstack((val_feature_1.reshape(-1,1), val_feature_2.reshape(-1,1)))
y_valid_features = y_valid
x_test_features = np.hstack((test_feature_1.reshape(-1,1), test_feature_2.reshape(-1,1)))
y_test_features = y_test
#x_train = x_train.reshape(-1,1)
#x_valid = x_valid.reshape(-1,1)
#x_test = x_test.reshape(-1,1)
plt.plot(x_train, y_train, 'ro')
plt.xlabel('X Value')
plt.ylabel('y Value')
plt.title('Univariate Training Dataset (Group-8)')
plt.show()
#print(x_train[0:5])
#print(y_train[0:5])
```

```python
clf = linear_model.SGDRegressor(alpha = 0, eta0=0.0003, fit_intercept=True, learning_rate = 'constant', loss = 'squared_loss',  max_iter=1000, penalty = 'l2', shuffle=True)
clf.fit(x_train_features, y_train_features)


```

```python
y_pred_val = clf.predict(x_valid_features)
#print(y_pred_val[0:10],y_valid[0:10])
loss_valid = np.mean(np.power((y_pred_val - y_valid_features),2))
print(loss_valid)
plt.scatter(y_valid_features, y_pred_val)
plt.xlabel('True Y value')
plt.ylabel('Predicted Y value')
plt.title('Scatter Plot for UNIVARIATE DATA on validation Data')
plt.show()
plt.plot(x_valid,y_valid_features, 'b^',x_valid, y_pred_val, 'ro')
plt.xlabel('X Value')
plt.ylabel('y Value')
plt.title('Univariate Validation Dataset (Group-8) along with the Predicted Function')
plt.show()
```

```python
y_pred_train = clf.predict(x_train_features)
#print(y_pred_val[0:10],y_valid[0:10])
loss_train = np.mean(np.power((y_pred_train - y_train_features),2))
print(loss_train)
plt.scatter(y_train_features, y_pred_train)
plt.xlabel('True Y value')
plt.ylabel('Predicted Y value')
plt.title('Scatter Plot for UNIVARIATE DATA on Training Data')
plt.show()
plt.plot(x_train,y_train_features, 'g^',x_train, y_pred_train, 'yo')
plt.xlabel('X Value')
plt.ylabel('y Value')
plt.title('Univariate Training Dataset (Group-8) along with the Predicted Function')
plt.show()
```

```python
y_pred_test = clf.predict(x_test_features)
loss_test = np.mean(np.power((y_pred_test - y_test_features),2))
print(loss_test)
plt.scatter(y_test_features, y_pred_test)
plt.xlabel('True Y value')
plt.ylabel('Predicted Y value')
plt.title('Scatter Plot for UNIVARIATE DATA on Test Data')
plt.show()
plt.plot(x_test,y_test_features, 'b.',x_test, y_pred_test, 'ro')
plt.xlabel('X Value')
plt.ylabel('y Value')
plt.title('Univariate Test Dataset (Group-8) along with the Predicted Function')
plt.show()
plt.plot(y_test_features, 'b.',y_pred_test, 'ro')
#plt.xlabel('X Value')
#plt.ylabel('y Value')
#plt.title('Univariate Test Dataset (Group-8) along with the Predicted Function')
plt.show()
```
