```python
import numpy as np
#import sklearn as sk
from sklearn import linear_model
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
import tensorflow as tf
import csv
```

```python
x_train = []
y_train = []
x_valid = []
y_valid = []
x_test = []
y_test = []

with open('./../bivariateData/group8_train.txt', 'r') as file:
    for row in file:
        a,b,c = row.split()
        x_train.append([float(a), float(b)])
        y_train.append(float(c))

with open('./../bivariateData/group8_val.txt', 'r') as file:
    for row in file:
        a,b,c = row.split()
        x_valid.append([float(a), float(b)])
        y_valid.append(float(c))
        
with open('./../bivariateData/group8_test.txt', 'r') as file:
    for row in file:
        a,b,c = row.split()
        x_test.append([float(a), float(b)])
        y_test.append(float(c))
        
x_train = np.array(x_train)
y_train = np.array(y_train)
x_valid = np.array(x_valid)
y_valid = np.array(y_valid)
x_test = np.array(x_test)
y_test = np.array(y_test)

train_feature_1 = (x_train[:,0] - np.mean(x_train[:,0]))/np.std(x_train[:,0])
train_feature_2 = (x_train[:,1] - np.mean(x_train[:,1]))/np.std(x_train[:,1])
x_train2 = np.multiply(x_train[:,0], x_train[:,1])
train_feature_3 = (x_train2 - np.mean(x_train2))/np.std(x_train2)
val_feature_1 = (x_valid[:,0] - np.mean(x_valid[:,0]))/np.std(x_valid[:,0])
val_feature_2 = (x_valid[:,1] - np.mean(x_valid[:,1]))/np.std(x_valid[:,1])
x_valid2 = np.multiply(x_valid[:,0], x_valid[:,1])
val_feature_3 = (x_valid2 - np.mean(x_valid2))/np.std(x_valid2)
test_feature_1 = (x_test[:,0] - np.mean(x_test[:,0]))/np.std(x_test[:,0])
test_feature_2 = (x_test[:,1] - np.mean(x_test[:,1]))/np.std(x_test[:,1])
x_test2 = np.multiply(x_test[:,0], x_test[:,1])
test_feature_3 = (x_test2 - np.mean(x_test2))/np.std(x_test2)


x_train_features = np.hstack((train_feature_1.reshape(-1,1), train_feature_2.reshape(-1,1), train_feature_3.reshape(-1,1)))
#print(x_train_features[0:5, :])
y_train_features = y_train
x_valid_features = np.hstack((val_feature_1.reshape(-1,1), val_feature_2.reshape(-1,1), val_feature_3.reshape(-1,1)))
y_valid_features = y_valid
x_test_features = np.hstack((test_feature_1.reshape(-1,1), test_feature_2.reshape(-1,1), test_feature_3.reshape(-1,1)))
y_test_features = y_test
#x_train = x_train.reshape(-1,1)
#x_valid = x_valid.reshape(-1,1)
#x_test = x_test.reshape(-1,1)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(list(x_train[:,0]), list(x_train[:,1]), list(y_train), zdir='z' )
ax.set_xlabel('X1 Value')
ax.set_ylabel('X2 Value')
ax.set_zlabel('Y Value')
ax.set_title('Biivariate Training Dataset (Group-8)')
plt.show()
#print(x_train[0:5])
#print(y_train[0:5])
```

```python
clf = linear_model.SGDRegressor(alpha = 0, eta0=0.0003, fit_intercept=True, learning_rate = 'constant', loss = 'squared_loss',  max_iter=1000, penalty = 'l2', shuffle=True)
clf.fit(x_train_features, y_train_features)


```

```python
y_pred_val = clf.predict(x_valid_features)
#print(y_pred_val[0:10],y_valid[0:10])
loss_valid = np.mean(np.power((y_pred_val - y_valid_features),2))
print(loss_valid)
plt.scatter(y_valid_features, y_pred_val)
plt.xlabel('True Y value')
plt.ylabel('Predicted Y value')
plt.title('Scatter Plot for BIVARIATE DATA on validation Data')
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(list(x_valid[:,0]), list(x_valid[:,1]), list(y_valid_features), zdir='z' )
ax.set_xlabel('X1 Value')
ax.set_ylabel('X2 Value')
ax.set_zlabel('Y Value')
ax.set_title('Bivariate Validation Dataset (Group-8) along with the Predicted Function')

ax.scatter(list(x_valid[:,0]), list(x_valid[:,1]), list(y_pred_val), zdir='z' )
plt.show()
```

```python
y_pred_train = clf.predict(x_train_features)
#print(y_pred_val[0:10],y_valid[0:10])
loss_train = np.mean(np.power((y_pred_train - y_train_features),2))
print(loss_train)
plt.scatter(y_train_features, y_pred_train)
plt.xlabel('True Y value')
plt.ylabel('Predicted Y value')
plt.title('Scatter Plot for BIVARIATE DATA on training Data')
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(list(x_train[:,0]), list(x_train[:,1]), list(y_train_features), zdir='z' )
ax.set_xlabel('X1 Value')
ax.set_ylabel('X2 Value')
ax.set_zlabel('Y Value')
ax.set_title('Bivariate Training Dataset (Group-8) along with the Predicted Function')

ax.scatter(list(x_train[:,0]), list(x_train[:,1]), list(y_pred_train), zdir='z' )
plt.show()
```
